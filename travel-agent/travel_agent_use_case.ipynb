````python
from workflow import app
from models import WorkflowState
from langchain_core.messages import HumanMessage
from services.markdown_exporter import MarkdownExporter
i=2
user_input = [
  "I want to go to Paris for 3 days, my budget is 1000 EUR, I like art and culture, my currency is USD",
  "Plan a 5-day trip to Tokyo for 2 people, budget 2000 USD, interested in food and anime",
  "Suggest a weekend in Rome, 500 EUR, history and architecture, currency is GBP"
][i]
state = WorkflowState(
  destination=None,
  budget=None,
  native_currency=None,
  days=None,
  group_size=None,
  activity_preferences=None,
  accommodation_type=None,
  dietary_restrictions=None,
  transportation_preferences=None,
  messages=[HumanMessage(content=user_input)]
)
final_state = app.invoke(state)

# Build the full markdown content
md = []
md.append(f"# ðŸ§³ AI Travel Agent: Full Reasoning\n")
md.append(f"**User Input:** {user_input}\n")
md.append("## 1. Query Analysis\n")
md.append(f"- Destination: {final_state['destination']}\n- Budget: {final_state['budget']}\n- Native Currency: {final_state['native_currency']}\n- Days: {final_state['days']}\n- Activity Preferences: {final_state['activity_preferences']}\n")
md.append("## 2. Hotel Search\n")
md.append("```python\n" + "\n".join(str(h) for h in final_state['hotels']) + "\n```\n")
md.append("## 3. Weather Forecast\n")
md.append(f"{final_state['weather']}\n")
md.append("## 4. Attractions\n")
md.append(f"{final_state['attractions']}\n")
md.append("## 5. Budget Calculation\n")
md.append(f"{final_state['calculator_result']}\n")
md.append("## 6. Itinerary\n")
md.append("```python\n" + str(final_state['itinerary']) + "\n```\n")
md.append("## 7. Final Summary\n")
md.append(final_state['summary']['summary'])

full_markdown = "\n".join(md)

exporter = MarkdownExporter(disclaimer=True)
filename = f"ai_travel_plan_full_trace_example_{i+1}.md"
exporter.export(full_markdown, filename)
print(f"Saved: {filename}")
````
